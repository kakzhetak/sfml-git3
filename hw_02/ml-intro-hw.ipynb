#### Задание 3. Фунция предобработки

Напишите функцию, которая бы

* Удаляло идентификатор `AGREEMENT_RK`
* Избавлялась от проблем с '.' и ',' в стобцах PERSONAL_INCOME, CREDIT, FST_PAYMENT, LOAN_AVG_DLQ_AMT, LOAN_MAX_DLQ_AMT
* Что-то делала с пропусками
* Кодировала категориальные признаки

В результате, ваш датафрейм должен содержать только числа и не содержать пропусков!

```python
def preproc_data(df_input):
    df_output = df_input.copy()
    df_output = df_output.drop(['AGREEMENT_RK'], axis=1)
    columns = ['PERSONAL_INCOME', 'CREDIT', 'FST_PAYMENT', 'LOAN_AVG_DLQ_AMT', 'LOAN_MAX_DLQ_AMT']
    for column in columns:
        df_output[column] = df_output[column].map(lambda x: x.replace(',', '.')).astype('float')
    
    df_output['TP_PROVINCE'] = df_output['TP_PROVINCE'].fillna(df_output['FACT_ADDRESS_PROVINCE'])
    df_output.loc[df_output['GEN_INDUSTRY'].isnull(),'GEN_INDUSTRY'] = 'Другое'
    df_output.loc[df_output['GEN_TITLE'].isnull(),'GEN_TITLE'] = 'Другое'
    df_output.loc[df_output['JOB_DIR'].isnull(),'JOB_DIR'] = 'Другое'
    df_output.loc[df_output['ORG_TP_STATE'].isnull(),'ORG_TP_STATE'] = 'Другое'
    df_output.loc[df_output['GEN_INDUSTRY'].isnull(),'GEN_INDUSTRY'] = 'Другое'
    df_output.loc[df_output['GEN_INDUSTRY'].isnull(),'GEN_INDUSTRY'] = 'Другое'
    df_output.loc[df_output['ORG_TP_FCAPITAL'].isnull(),'ORG_TP_FCAPITAL'] = 'Без участия'
    df_output.loc[df_output['WORK_TIME'].isnull(),'WORK_TIME'] = df_output['WORK_TIME'].mean()
    
    ## Your Code Here
    
    return df_output
df_preproc = df.pipe(preproc_data)
```
